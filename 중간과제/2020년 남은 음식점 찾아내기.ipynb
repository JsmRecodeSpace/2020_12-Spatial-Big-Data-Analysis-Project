```python
df = pd.read_csv('F&B_Store_201012_except_pubs.csv', encoding='cp949')

foodDF = df.drop(['wkt_geom', 'UPJONG1_NM',  '주소구분','ADDRESS','우편번호','PNU','CLSS','X','Y','PHONE', '입력주소','표준신주소','표준구주소'], axis=1)

foodDF = foodDF.rename(columns={'UPJONG2_NM':'upjong2_nm', 'UPJONG3_NM':'upjong3_nm','STORE_NM':'store_nm'})

foodDF.head()
```

### 네이버에서 검색하여 보기

```python
driver_path = 'C:\\Users\\82105\\Downloads\\chromedriver_win32\\chromedriver'

for i in range(300, len(store_name_lst)):
    
    # chrome 작동
    driver = webdriver.Chrome(driver_path)
    
    # 구버전 네이버지도 접속
    driver.get("https://v4.map.naver.com")
    
    # 네이버 지도 업데이트 후 안내메시지 끄기
    driver.find_elements_by_css_selector("button.btn_close")[1].click()
    time.sleep(1)
    
    
    # 검색창에 검색어 입력하기 // 검색창: input#search-input
    search_box = driver.find_element_by_css_selector("input#search-input")
    search_box.send_keys(f'{store_name_lst[i]}')
    # 검색버튼 누르기 // 검색버튼: button.spm
    search_button = driver.find_element_by_css_selector("button.spm")
    search_button.click()
    time.sleep(3)
    
    driver.close()

    
```
